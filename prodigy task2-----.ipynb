```python
#data cleaning and preprocessing
import pandas as pd

# Fill missing values in 'Age' with the median age for both train and test
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)

# Handle missing values in 'Embarked'
if 'Embarked' in train_data.columns:
    if train_data['Embarked'].isnull().sum() > 0:
        # Calculate mode
        mode_embarked = train_data['Embarked'].mode()
        if not mode_embarked.empty:
            # Fill missing values in 'Embarked' with the most common port
            train_data['Embarked'].fillna(mode_embarked[0], inplace=True)
        else:
            print("Mode calculation for 'Embarked' returned empty result.")
    else:
        print("'Embarked' has no missing values.")
else:
    print("Column 'Embarked' not found in train_data.")

# Fill missing values in 'Fare' in the test data with the median fare
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)

# Drop 'Cabin' column from both datasets if it exists
if 'Cabin' in train_data.columns:
    train_data.drop('Cabin', axis=1, inplace=True)
else:
    print("Column 'Cabin' not found in train_data.")

if 'Cabin' in test_data.columns:
    test_data.drop('Cabin', axis=1, inplace=True)
else:
    print("Column 'Cabin' not found in test_data.")

```

```python
# Encode 'Sex' column
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})
train_data
test_data
# Encode 'Embarked' column
train_data['Embarked'] = train_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
test_data['Embarked'] = test_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

train_data
test_data

# Create 'FamilySize' and 'IsAlone' features
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch']
train_data['IsAlone'] = (train_data['FamilySize'] == 0).astype(int)
train_data
test_data

test_data['FamilySize'] = test_data['SibSp'] + test_data['Parch']
test_data['IsAlone'] = (test_data['FamilySize'] == 0).astype(int)

train_data
test_data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

# Prepare features and target variable
# Drop 'PassengerId' as it's not useful for modeling
X = train_data.drop(['Survived', 'PassengerId'], axis=1)
y = train_data['Survived']

# Handle categorical features
# Convert 'Sex' to numerical values
X['Sex'] = X['Sex'].map({'male': 0, 'female': 1})

# One-hot encode 'Embarked' column if not already done
if 'Embarked' in X.columns:
    X = pd.get_dummies(X, columns=['Embarked'], drop_first=True)

# Handle missing values separately for numeric and non-numeric columns
# Identify numeric columns
numeric_cols = X.select_dtypes(include=['number']).columns
# Fill missing values in numeric columns with median
X[numeric_cols] = X[numeric_cols].fillna(X[numeric_cols].median())

# Identify non-numeric columns and handle them if necessary
non_numeric_cols = X.select_dtypes(exclude=['number']).columns
for col in non_numeric_cols:
    if X[col].dtype == 'object':
        mode_value = X[col].mode()[0]
        X[col].fillna(mode_value, inplace=True)

# Ensure that all data used in modeling is numeric
X = pd.get_dummies(X, drop_first=True)

# Split the training data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize and train the model
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)

# Make predictions on the validation set
y_pred = model.predict(X_valid)

# Evaluate the model
accuracy = accuracy_score(y_valid, y_pred)
print(f'\nValidation Accuracy: {accuracy:.4f}')
```
